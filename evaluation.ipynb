```python
import run_mcstas as rms
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("/home/cherb/LRZSync/Doktorarbeit/Vorlagen_Verschiedenes/stylelibs/christoph.mplstyle")

sim = rms.McstasSimulation("/home/cherb/LRZSync/Doktorarbeit/Mieze/Logarithmic_Spiral/mcstas_component/", "reverse_test.instr")
sim.init_params_dict()
sim.new_compile = False
sim.num_neutrons=1e6
pd = sim.return_params_dict()
data_dict = sim.run_and_return()
data_dict = sim.return_images_data()
images_dict = sim.plot_last_images()
all_ratios = []
#sim.delete_last_folder()
for source_width in np.linspace(0.00001, 0.006, 13):
    pd['source_width'] = source_width
    data_dict = sim.run_and_return()
    data_dict = sim.return_images_data()
    images_dict = sim.plot_last_images()
    intdata = np.sum(data_dict['psdafterlog.dat'][:500, :], axis=0)
    fig, ax = plt.subplots(1)
    ax.plot(np.linspace(-5, 5 , 500), intdata, linestyle='-', marker=' ')
    def mm2pixel(mm):
        return int((mm+5)*500/10)
    ax.vlines([-0.3, 0.3], 0, np.max(intdata))
    wrongratio = np.sum(intdata[mm2pixel(-0.4): mm2pixel(0.4)])/np.sum(intdata)
    print(mm2pixel(-0.3), wrongratio)

    ax.set_xlabel("$x$ (cm)")
    ax.set_ylabel("int. intensity (arb. u.)")
    ax.set_title("ratio of wrong neutrons {:.2}".format(wrongratio))
    all_ratios.append(wrongratio)
print(all_ratios)
```

```python
plt.style.use("/home/cherb/LRZSync/Doktorarbeit/Vorlagen_Verschiedenes/stylelibs/christoph.mplstyle")
fig, ax = plt.subplots(1)
ax.plot(np.linspace(0.00001, 0.006, 13)*1000, all_ratios, marker='s', linestyle=' ')
ax.set_xticks(list(range(7)))
ax.set_xlabel("source width $w$ (mm)")
ax.set_ylabel("wrong neutrons/all neutrons")
fig.savefig("/home/cherb/LRZSync/Doktorarbeit/leakagevsw.pdf", bbox_inches='tight')
```
